```python
customerchurnpred['gender']=customerchurnpred['gender'].astype('category').cat.codes
```

```python
x=customerchurnpred.drop(['churn'],axis=1).values
```

```python
y=customerchurnpred['churn'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,BaggingClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
models={
    "DecisionTree":DecisionTreeClassifier(),
    "RandomForest":RandomForestClassifier(),
    "GradientBoosting":GradientBoostingClassifier(),
    "BC":BaggingClassifier()
}
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    print(f"{name}:Accuracy={accuracy:.6f}")
    classification=classification_report(y_test,y_pred)
    print(f"{name}\n{classification}\n")
```
